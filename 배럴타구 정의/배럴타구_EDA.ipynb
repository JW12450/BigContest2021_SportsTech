```python
#비율 누적
ratio2['ratio_accum'] = 0.0
ratio2['ratio_accum'][0] = ratio2['ratio'][0]

for i in range(len(ratio2)-1):
  ratio2['ratio_accum'][i+1] = ratio2['ratio_accum'][i] + ratio2['ratio'][i+1]

ratio2
```
